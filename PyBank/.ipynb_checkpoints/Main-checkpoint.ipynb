```python
# -*- Unit 2 Homework PyBank -*-

# Import CSV File and create file path
from pathlib import Path
import csv

# set path of the files
csvpath = Path("/Users/riteshchugani/Desktop/Unit2HW/PyBank/budget_data.csv")

# Initialize list of data points to calculate and declare variables
Months_List = []
Total_Months = 0

Profit_Losses_List = []
Total_Profit_Losses = 0

# Open the csv file and make sure it is the correct type (object)
with open(csvpath, "r") as csvfile:
    print(type(csvfile))
    
    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=",")
    
    # Read the header row first (skip this step if there is no header)
    csv_header = next(csvreader)
    
    # Read each row of data after the header
    for row in csvreader:
        # Set the ‘Total_Months’ variable and calculate total number of months included in the dataset
        Total_Months += 1
                
        # Set Profit/Loss variable and calculate net total amount of Profit/Losses over the entire period
        Profit_Loss = int(row[1])
        Total_Profit_Losses += Profit_Loss
        
        # Update the list of Months and Monthly Profit/Losses
        Months_List.append(row[0])
        Profit_Losses_List.append(Profit_Loss)
        
        # Calculating monthly profit and loss difference
        x = Profit_Losses_List
        # Calculating difference between adjacent items in a python list - in this case the monthly profit/losses
        Monthly_Differences = [x[n]-x[n-1] for n in range(1,len(x))] 
        #Adjusting the 1st month value as there is no value and also to have the same number of values of month and monthly change
        Monthly_Differences.insert(0,0) 
        
        # Calculating average of the changes in Profit/Losses over the entire period
        a = sum(Monthly_Differences)
        b = len(Monthly_Differences[1:])
        c = 0 if a == 0 else (a/b) #remove 0 error due to difference in len
        Monthly_avg = round(c,2)

#The greatest increase in profits ( amount and month ) over the entire period
greatest_increase_pnl = max(Monthly_Differences)
greatest_corresponding_profit = Profit_Losses_List[Monthly_Differences.index(greatest_increase_pnl)]
greatest_increase_month = Months_List[Profit_Losses_List.index(greatest_corresponding_profit)]

#The greatest decrease in profits ( amount and month ) over the entire period
greatest_decrease_pnl = min(Monthly_Differences)
greatest_corresponding_loss = Profit_Losses_List[Monthly_Differences.index(greatest_decrease_pnl)]
greatest_decrease_month = Months_List[Profit_Losses_List.index(greatest_corresponding_loss)]

print(str("--------------------"))
print(str("Financial Analysis"))
# Total number of months included in the csv
print(f"Total Months: {Total_Months} ")
# The net total amount of profit and losses over the entire period
print(f'Total: ${Total_Profit_Losses}')
# The average changes in P/L
print(f'Average Change: {Monthly_avg}')
# The greatest increase in profits (date and amount) over the entire period
print(f'Greatest Increase in Profits: {greatest_increase_month} (${greatest_increase_pnl})')
# The greatest decrease in losses (date and amount) over the entire period
print(f'Greatest Decrease in Profits: {greatest_decrease_month} (${greatest_decrease_pnl})')
print(str("--------------------"))


# Set output file name
output_path = 'output.txt'

# Open the output path as a file object
with open(output_path, 'w') as file:
    # Write daily_average to the output file, convert to string
    file.write(f"Financial Analysis\n")
    file.write("---------------------\n")
    file.write(f"Total Months: {Total_Months}  \n ")
    file.write(f'Total: ${Total_Profit_Losses} \n')
    file.write(f'Average Change: {Monthly_avg} \n')
    file.write(f'Greatest Increase in Profits: {greatest_increase_month} (${greatest_increase_pnl})\n')
    file.write(f'Greatest Decrease in Profits: {greatest_decrease_month} (${greatest_decrease_pnl})\n')
    file.write("---------------------\n")
```
